#### Put data from python

###### First example: 1D array

```python
import numpy as np
b = {}
b['name']='myvar'                         
x=np.linspace(-np.pi*4, np.pi*4, 500)
b['data']=np.sin(x)/x
b.keys()
```

```python
%ferret_putdata --axis_pos (0,1,2,3,4,5) b
```

###### Second example: 3D array

Create a dummy 3D array

```python
ndepths, nlats, nlons = (10, 73, 145)

lats = np.linspace(-np.pi / 2, np.pi / 2, nlats)
lons = np.linspace(0, 2 * np.pi, nlons)
lons, lats = np.meshgrid(lons, lats)

wave = 0.75 * (np.sin(2 * lats) ** 8) * np.cos(4 * lons)
mean = 0.5 * np.cos(2 * lats) * ((np.sin(2 * lats)) ** 2 + 2)

lats = np.rad2deg(lats)
lons = np.rad2deg(lons)
data2D = wave + mean 

depths = np.linspace(1, 1000, ndepths)
data3D = np.repeat(np.expand_dims(data2D,axis=0), ndepths, axis=0)

print data3D.shape
```

Please refer to http://ferret.pmel.noaa.gov/Ferret/documentation/pyferret/data-dictionaries/

```python
import pyferret
data2ferret = {}
data2ferret['name']='myvar'   
data2ferret['axis_names']=('depth','lats', 'lons')
data2ferret['axis_types']=(pyferret.AXISTYPE_LEVEL, pyferret.AXISTYPE_LATITUDE, pyferret.AXISTYPE_LONGITUDE)
data2ferret['axis_units']=('meters','degrees_north','degrees_east')
data2ferret['axis_coords']=(depths[:], lats[:,0], lons[0,:])
data2ferret['data']=data3D
data2ferret.keys()
```

```python
#%ferret_putdata --axis_pos (0,1,2,3,4,5) lon2ferret
#%ferret_putdata --axis_pos (1,0,2,3,4,5) lats2ferret
%ferret_putdata data2ferret
```
