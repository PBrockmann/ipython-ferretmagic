#### Put data from python

```python
import numpy as np
b = {}
b['name']='myvar'                         
x=np.linspace(-np.pi*4, np.pi*4, 500)
b['data']=np.sin(x)/x
b.keys()
```

```python
%ferret_putdata --axis_pos (1,0,2,3,4,5) b
```

```python
%%ferret -a
show data
show grid myvar
ppl color 2, 0, 50, 100
ppl color 3, 100, 50, 0
plot/thick=3/color myvar, myvar[y=@shf:50]
```
