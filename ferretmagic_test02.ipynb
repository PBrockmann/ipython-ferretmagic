#### Put data from python

###### First example: 1D array

```python
import numpy as np
b = {}
b['name']='myvar'                         
x=np.linspace(-np.pi*4, np.pi*4, 500)
b['data']=np.sin(x)/x
b.keys()
```

```python
%ferret_putdata --axis_pos (0,1,2,3,4,5) b
```

```python
%%ferret
use levitus_climatology    ! A dataset must have been opened before putting data to ferret
show grid myvar
ppl color 2, 0, 50, 100
ppl color 3, 100, 50, 0
plot/thick=3/color myvar, myvar[x=@shf:50]
```

###### Second example: 2D array

```python
nlats, nlons = (73, 145)
lats = np.linspace(-np.pi / 2, np.pi / 2, nlats)
lons = np.linspace(0, 2 * np.pi, nlons)
lons, lats = np.meshgrid(lons, lats)
wave = 0.75 * (np.sin(2 * lats) ** 8) * np.cos(4 * lons)
mean = 0.5 * np.cos(2 * lats) * ((np.sin(2 * lats)) ** 2 + 2)

lats = np.rad2deg(lats)
lons = np.rad2deg(lons)
data = wave + mean
lons.shape
```

```python
data2ferret = {}
data2ferret['name']='myvar'   
data2ferret['axis_names']=('lats','lons')
data2ferret['axis_types']=(2,1)
data2ferret['axis_units']=('degrees_north','degrees_east')
data2ferret['axis_coords']=(lats[:,0],lons[0,:])
data2ferret['data']=data
data2ferret.keys()
```

```python
#%ferret_putdata --axis_pos (0,1,2,3,4,5) lon2ferret
#%ferret_putdata --axis_pos (1,0,2,3,4,5) lats2ferret
%ferret_putdata data2ferret
```
